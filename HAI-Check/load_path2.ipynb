```python
load_path ="./hai/HAI-Check/data/외부자소서DB_20240411.xlsx"
def make_load_path2_df(load_path):
    formats = load_path.split('.')[-1]
    if formats == 'xlsx':
        loaded_df = pd.read_excel(load_path, engine='openpyxl').fillna(' ').astype(str)
    elif formats == 'xls':
        loaded_df = pd.read_excel(load_path).fillna(' ').astype(str)
    df2 = pd.DataFrame()
    df2['수험번호'] = loaded_df['수집한 사이트'] + loaded_df['no']
    df2['지원분야'] = loaded_df['직렬']
    df2['지원자명'] = loaded_df['no']
    df2['자기소개서1'] = loaded_df['내용']
        
    return df2
```
