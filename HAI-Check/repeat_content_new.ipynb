```python
for i in df.columns:
    df.loc[:,i] = df.loc[:,i].str.split('.')
```

```python
arg = 5

for user in range(len(df)):
    for docs in df.columns:
        target = [s.strip() + "." for s in df.loc[user,docs]]
        for sent in target:
            if len(sent.split(" ")) > arg:
                count = target.count(sent)
                if count > 1:
                    print(loaded_df.loc[user, '수험번호'], docs, sent, count)
                    print()
        
```
